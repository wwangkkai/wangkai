# 循环

## while 循环
- 当一个条件保持真的时候while循环重复执行语句
- while 循环一定要有结束条件，否则很容易进入死循环
- while 循环的语法是：

    while loop-contunuation-conndition:

        Statement

## 示例：
sum = 0

i = 1

while i <10:

    sum = sum + i
    i = i + 1

## 错误示例：
sum = 0

i = 1

while i <10:

    sum = sum + i

i = i + 1
- 一旦进入死循环可按 Ctrl + c 停止

# 验证码
- 随机产生四个字母的验证码，如果正确，输出验证码正确。如果错误，产生新的验证码，用户重新输入。
- 验证码只能输入三次，如果三次都错，返回“别爬了，我们小网站没什么好爬的”
- 密码登录，如果三次错误，账号被锁定


```python
import random
i = 0
while i <3 :
    yanzhengma = str(random.randint(1000,9999))
    print('验证码是:',yanzhengma)
    input_ = input('>>')
    if yanzhengma == input_:
        print('验证码正确')
        break
    else:
        print('请再输入')
        i +=1
#     if i ==3:
#         print('验证码次数超限')
else:
    print('验证码次数超限')
```

## 嵌套循环
- 一个循环可以嵌套另一个循环
- 每次循环外层时，内层循环都会被刷新重新完成循环
- 也就是说，大循环执行一次，小循环会全部执行一次
- 注意：
> - 多层循环非常耗时
  - 最多使用3层循环

```python
for i in range(10):        #也就是说，大循环执行一次，小循环会全部执行一次
    for j in range(10):
        for k in range(10):
            print(i,j,k)
```
